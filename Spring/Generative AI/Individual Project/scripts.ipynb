```python
import pandas as pd

# Example of creating a DataFrame with intraday data

df = pd.read_csv("timeseries_TSLA_2025-04-21.csv")

# Calculate daily return
df['return'] = df['close'].pct_change()

# Calculate moving averages
df['SMA_5'] = df['close'].rolling(window=5).mean()  # 5-period simple moving average
df['SMA_10'] = df['close'].rolling(window=10).mean()  # 10-period simple moving average

# Summary statistics
summary_stats = df.describe()

# Display results
print(summary_stats)
```
