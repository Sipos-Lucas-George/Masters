```python
import nltk
import numpy as np
from nltk.corpus import names
import random
```

```python
def gender_features(word):
    return {
        'last_letter': word[-1],
        # 'last_two_letters': word[-2:],  # New feature
        # 'length': len(word)  # New feature
    }
# acc = []
# for _ in range(1000):
male_names = [(name, 'male') for name in names.words('male.txt')]
female_names = [(name, 'female') for name in names.words('female.txt')]
labeled_names = male_names + female_names
random.shuffle(labeled_names)

featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]
train_set, test_set = featuresets[500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)

ans1 = classifier.classify(gender_features('Mark'))
ans2 = classifier.classify(gender_features('Precilla'))

print("Mark is:", ans1)
print("Precilla is:", ans2)

classifier.show_most_informative_features()
print(len(classifier.most_informative_features()))

# Checking the accuracy on the test set
acc.append(nltk.classify.accuracy(classifier, test_set))
# print("Accuracy:", acc[-1])
print(np.mean(acc))
# last letter
# 0.761854
# 0.760038

# last 2 letters
# 0.780704
# 0.783168
```
