```python
from collections import Counter
```

```python
text = "sky car car sky sky sky sky sky sky sky sky sky hello hello hello hello hello hello hello bye bye bye bye bye bye why why why why why apple apple apple apple hi hi hi hi a a a a a a a a a a mine mine mine mine mine work".split(" ")
word_counter = Counter(text)
number_of_words = len(text)
print(f"Number of words in text: {number_of_words}")
print(word_counter)
```

Now, I'll get the frequency of the words over the total number of words in text, without the words that are ***not included***

```python
excluded_words = ["a", "why", "hi"]
frequency = []
for word, freq in word_counter.items():
    if word in excluded_words:
        number_of_words -= freq
        continue
    frequency.append((word, freq))
frequency = [(word, f/number_of_words) for word, f in frequency]
print(f"Number of words in text without excluded: {number_of_words}")
print(f"Frequency of words in text: {sorted([(w, f) for w, f in word_counter.items() if w not in excluded_words], key=lambda x: -x[1])}")
print(f"Frequency of words in text: {sorted(frequency, key=lambda x: -x[1])}")
print(f"Minimum frequency to be accepted in wordcloud: {sum([f for w, f in frequency])/len(frequency)}")
```
