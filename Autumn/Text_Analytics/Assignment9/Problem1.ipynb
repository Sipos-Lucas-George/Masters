```python
import numpy as np
import pandas as pd
from sklearn.metrics import cohen_kappa_score

ratings = {
    "Opinion": [
        "Battery lasts all day.",
        "Camera not as good as expected.",
        "Okay for basic use.",
        "Screen resolution is fantastic.",
        "Too expensive.",
        "Apps crash.",
        "Lightweight and easy to carry.",
        "Software confusing.",
        "Decent but not exceptional.",
    ],
    "Rater 1": ["P", "N", "O", "P", "N", "N", "P", "N", "O"],
    "Rater 2": ["P", "N", "O", "P", "N", "N", "P", "O", "O"],
    "Rater 3": ["O", "N", "P", "P", "N", "N", "P", "O", "O"],
}

df = pd.DataFrame(ratings)

rating_map = {"P": 1, "O": 0, "N": -1}
df_numeric = df.replace(rating_map).iloc[:, 1:]

# Calculate pairwise Cohen's Kappa
kappa_12 = cohen_kappa_score(df["Rater 1"], df["Rater 2"])
kappa_13 = cohen_kappa_score(df["Rater 1"], df["Rater 3"])
kappa_23 = cohen_kappa_score(df["Rater 2"], df["Rater 3"])
print(*[("kappa_12", kappa_12), ("kappa_13", kappa_13), ("kappa_23", kappa_23)], sep="\n")

average_kappa = np.mean([kappa_12, kappa_13, kappa_23])

# Calculate Pearson correlation
correlation_matrix = df_numeric.corr()

print("Kappa avg:", average_kappa) 
print(correlation_matrix)
```
