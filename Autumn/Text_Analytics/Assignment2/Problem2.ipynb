```python
import nltk
from nltk import PorterStemmer, WordNetLemmatizer
```

```python
with open('file_problem_2') as file:
    text = file.readlines()[0]
text
```

```python
tokenized = nltk.word_tokenize(text)
print(tokenized)
```

```python
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in tokenized]
print(stemmed_tokens)
```

```python
pos_tags = nltk.pos_tag(tokenized)
print(pos_tags)
```

```python
def get_pos_as_char(pos_tag: str) -> chr:
    if pos_tag[0] in "VRN":
        return pos_tag[0].lower()
    elif pos_tag[0] == 'J':
        return 'a'
    return 'n'
        
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token, get_pos_as_char(pos_tag)) for token, pos_tag in pos_tags]
print(lemmatized_tokens)
```
