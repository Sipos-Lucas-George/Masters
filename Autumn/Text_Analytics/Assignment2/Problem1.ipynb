```python
import nltk
```

```python
with open('file_problem_1') as file:
    text = file.readlines()[0]
text
```

```python
tokenized = nltk.word_tokenize(text)
print(tokenized)
```

```python
normalized = [word.lower() for word in tokenized]
print(normalized)
```

```python
pos_tags = nltk.pos_tag(normalized)
print(pos_tags)
```
