```python
def jaccard_distance(set1: set, set2: set) -> float:
    return round(1 - float(len(set1 & set2)) / len(set1 | set2), 2)
```

```python
def dice_coefficient(set1: set, set2: set) -> float:
    return 2 * float(len(set1 & set2)) / (len(set1) + len(set2))
```

```python
fruits = {
    'Apple': {"red", "sweet", "round", "juicy", "fruit"},
    'Banana': {"yellow", "sweet", "long", "peel", "fruit"},
    'Orange': {"orange", "round", "juicy", "citrus", "fruit"},
    'Grape': {"purple", "small", "juicy", "peel", "fruit"},
    'Lemon': {"yellow", "sour", "citrus", "peel", "fruit"}
}
jaccard_matrix = [[0]*5 for _ in range(5)]
dice_matrix = [[0]*5 for _ in range(5)]

fruit_names = list(fruits.keys())
```

```python
for i in range(5):
    for j in range(5):
        set1 = fruits[fruit_names[i]]
        set2 = fruits[fruit_names[j]]
        jaccard_matrix[i][j] = jaccard_distance(set1, set2)
        dice_matrix[i][j] = dice_coefficient(set1, set2)
```

```python
print("Jaccard Distance Matrix:")
for row in jaccard_matrix:
    print(row)

print("\nDice Coefficient Matrix:")
for row in dice_matrix:
    print(row)

print("\nChecking Triangle Inequality for Jaccard Distance:")
for i in range(5):
    for j in range(5):
        for k in range(5):
            if jaccard_matrix[i][k] > jaccard_matrix[i][j] + jaccard_matrix[j][k]:
                print(f"Triangle inequality failed for {fruit_names[i]}, {fruit_names[j]}, {fruit_names[k]}")
```
