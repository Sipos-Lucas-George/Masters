## Task 1

Download the World Happiness Index data from the link below, and load the data into a Pandas DataFrame.

http://mlg.ucd.ie/modules/COMP41680/happiness2019.csv

```python
import urllib.request

import pandas as pd
from matplotlib import pyplot as plt

url = "http://mlg.ucd.ie/modules/COMP41680/happiness2019.csv"
response = urllib.request.urlopen(url)
raw_csv = response.read().decode()
with open('happiness2019.csv', 'w') as csvfile:
    csvfile.write(raw_csv)
```

```python
data = pd.read_csv('happiness2019.csv', index_col=0)
data
```

```python
fig = plt.figure(figsize=(14, 7))
col_mean = sorted(data.mean(axis=1), reverse=True)
col_std = sorted(data.std(axis=1), reverse=True)
ax = fig.add_subplot(1, 2, 1)
ax.boxplot(col_mean[:5])
ax = fig.add_subplot(1, 2, 2)
ax.boxplot(col_std[:5])
plt.show()
```
